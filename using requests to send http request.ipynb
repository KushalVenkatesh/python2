```python
import http.client

conn = http.client.HTTPConnection("www.google.com")

headers = {
    'cache-control': "no-cache",
    'postman-token': "183f4fd3-6b39-e88b-3466-d45bf413852a"
    }

conn.request("GET", "/", headers=headers)

res = conn.getresponse()
data = res.read()

print(data.decode("utf-8"))
```

```python
import requests

url = "http://api.openweathermap.org/data/2.5/forecast"

querystring = {"zip":"95050","appid":"d50b273a083f6bcbfd5c74f892b74793","units":"imperial"}

headers = {
    'appid': "d50b273a083f6bcbfd5c74f892b74793",
    'zip': "95050",
    'units': "imperial",
    'Cache-Control': "no-cache",
    'Postman-Token': "01e48ea6-65c6-4cb0-9d6c-ad317d56ec7b"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```
